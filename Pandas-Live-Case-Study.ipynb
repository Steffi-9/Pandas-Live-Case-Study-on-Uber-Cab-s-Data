```python
1.Import Libraries

import pandas as pd

import numpy as np

2.Load Data

df=pd.read_csv("My Uber Drives.csv")

3.Read first 5 rows

df.head()

4.Read last rows

df.tail(7)

4.Get the total number of rows and columns

df.shape

print(f"The total no: of rows are {df.shape[0]} and columns are {df.shape[-1]}")

5.Get the total number of elements in the dataset

df.size

1156*7

6.Get the total no: of NULL values across every column in the dataset

df.isnull().sum()

df.isna().sum()

7.Get the Non-NULL values of every column

df.count()

df.info()

df.columns

8.Get the entries havinh NULL values in 'Purpose Column'

df['PURPOSE*'].isnull().sum()

df[df['PURPOSE*'].isnull()]

9.Get the entries having Non-Null values in the 'Purpose*' column

df[~df['PURPOSE*'].isnull()]

10.Remove the * in every column name using the rename function

df.rename(columns={'START_DATE*':'START_DATE', 'END_DATE*':'END_DATE', 'CATEGORY*':'CATEGORY', 'START*':'START', 'STOP*':'STOP', 'MILES*':'MILES',
       'PURPOSE*':'PURPOSE'})

df.head()

11.Remove the * in every column name using the str replace fn

df.columns.str.replace('*','')



12.Remove the * in every column using the lambda function

df.rename(mapper= lambda x: x.replace("*",""),axis=1,inplace=True)

df.columns

13.Get the entries in the data where the START location is 'Fort Pierce'

df[df['START']=='Fort Pierce']

14.Get the entries in data where the STOP location is 'Fort Pierce'

df[df['STOP']=='Fort Pierce']

15. Sort the entries in the data in descending order of the 'MILES' Column

df.sort_values(by='MILES',ascending=False)

16.Drop all the rows where there are NULL values in the STOP column

df=df.dropna(subset=['STOP'])

df[df['STOP'].isnull()]

17.Use the describe fn to get the statistical properties about the numerical columns in the data

df.describe()

18.Create a report in the html file using Pandas Profiling

from ydata_profiling import ProfileReport

# Generate the profiling report
profile = ProfileReport(df, title="Pandas Profiling Report", explorative=True)

# Save the report to an HTML file
profile.to_file("profiling_report.html")

19.Understanding the START and STOP points

print(df["START"].unique())

print(df["STOP"].unique())

len(df["START"].unique())

20.Use value_counts() fn demonstrate the proportion of different categorical values in the data

print(df["START"].value_counts())

df["STOP"].value_counts()

21.Get the no: of rides where START and STOP locations are the same

df[df["START"] == df["STOP"]][["START","STOP"]]

len(df[df["START"] == df["STOP"]][["START","STOP"]])

22.Find the favorite starting point according to the total no: of MILES covered

df.groupby(by="START").MILES.sum()

23.Find the starting point for the ride where maximum miles are covered

df[df["MILES"] == max(df["MILES"])][["START","MILES"]]

24.Check the data type of all the columns in the data set

df.dtypes

25.Drop the 'unknown location' value from START and STOP column

# Here we exclude the unknown location
df2=df[df["START"]!="Unknown Location"]
df2=df2[df2["STOP"]!="Unknown Location"]

df2

26.Find the most popular START-STOP pair according to the total no: of rides covered

df2.groupby(by=["START","STOP"]).size().sort_values(ascending=False).reset_index(name="Count")

27. Convert the datatypes of START_DATE and END_DATE column to datetime

df.dtypes


df["START_DATE"]=pd.to_datetime(df["START_DATE"])
df["END_DATE"]=pd.to_datetime(df["END_DATE"])
df.dtypes

28.Extract the month from START_DATE and try to get the proportion of rides of different months

df["Month"]=df["START_DATE"].dt.month

df.head(1)

df["Month_Name"]=df["START_DATE"].dt.month_name()

df.head(1)

df[["Month","Month_Name"]].value_counts().reset_index(name="Count")

29.Find the average distance covered each month

df.groupby("Month").MILES.mean().reset_index(name="MILES")

df.groupby("Month").mean(numeric_only=True)

30.Extract the day from the START_DATE column

df["DAY"]=df["START_DATE"].dt.day

df.head(1)

31.Find the total miles covered per category per purpose

df.groupby(["PURPOSE","CATEGORY"]).sum(numeric_only=True)["MILES"].sort_values(ascending=False).reset_index(name="Total Miles")

```
